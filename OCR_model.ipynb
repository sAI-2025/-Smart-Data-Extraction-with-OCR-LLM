```python
import cv2
import easyocr
import matplotlib.pyplot as plt

# Function to perform OCR
def perform_ocr(image_path):
    # Initialize the EasyOCR reader
    reader = easyocr.Reader(['en'], gpu=False)  # Add more languages like 'en', 'fr', etc.

    # Read the image using OpenCV
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError(f"Image at path {image_path} could not be loaded.")

    # Display the image (optional)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    # Perform OCR on the image
    ocr_result = reader.readtext(image)

    # Extract and return the text
    extracted_text = ''
    for detection in ocr_result:
        extracted_text += detection[1] + ' '  # detection[1] contains the text

    return extracted_text

```

```python
# @title
import easyocr
import os
import pandas as pd
import math

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'], gpu=False)

# Function to extract text from an image
def extract_text(image_path):
    # Read the image and get the OCR results
    result = reader.readtext(image_path)

    # Extract the text from the result
    extracted_text = ' '.join([text[1] for text in result])  # Extract the text part from each result
    return extracted_text

# Function to process images in batches and save results in a CSV file
def save_text_to_csv(folder_path, csv_file, batch_size=10):
    # Create a list to store image names and extracted text
    data = []

    # Get all image files from the specified folder (filtering for common image formats)
    image_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

    # Determine the total number of batches
    total_batches = math.ceil(len(image_paths) / batch_size)

    for batch_num in range(total_batches):
        # Get the current batch of image paths
        batch_paths = image_paths[batch_num * batch_size: (batch_num + 1) * batch_size]

        # Process each image in the current batch
        for image_path in batch_paths:
            # Extract the image name from the path
            image_name = os.path.basename(image_path)

            # Extract the text from the image
            extracted_text = extract_text(image_path)

            # Append the image name and extracted text to the data list
            data.append([image_name, extracted_text])

        print(f"Processed batch {batch_num + 1} of {total_batches}")

    # Create a DataFrame to store the data
    df = pd.DataFrame(data, columns=['Image Name', 'Extracted Text'])

    # Save the DataFrame to a CSV file
    df.to_csv(csv_file, index=False)

    print(f"Extracted text saved to {csv_file}")
    return df

```

```python
import pytesseract
from PIL import Image, ImageFile
import os
import pandas as pd
import gc  # For garbage collection to free up memory

# To avoid truncated image errors
ImageFile.LOAD_TRUNCATED_IMAGES = True

# Function to extract text from an image
def extract_text(image_path):
    try:
        # Open the image
        img = Image.open(image_path)

        # Extract text using pytesseract
        text = pytesseract.image_to_string(img)
        return text

    except OSError as e:
        # Handle the case where the image is corrupted or cannot be opened
        print(f"Error processing image {image_path}: {e}")
        return ""  # Return an empty string if the image is corrupted

# Function to process images in batches and save results in a CSV file
def save_text_to_csv(folder_path, csv_file, batch_size=100, skip_list=None):
    # Initialize skip_list if it's None
    if skip_list is None:
        skip_list = []

    # Create a list to store image names and extracted text
    data = []

    # Get all image files from the specified folder (filtering for common image formats)
    image_paths = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]

    # Open the CSV file in append mode so that we save in chunks
    with open(csv_file, 'a', newline='', encoding='utf-8') as f:
        for i in range(0, len(image_paths), batch_size):
            # Create a list to store image names and extracted text for the current batch
            batch_data = []

            # Process the current batch of images
            batch_paths = image_paths[i:i+batch_size]
            for image_path in batch_paths:
                # Extract the image name from the path
                image_name = os.path.basename(image_path)

                # Check if the image is in the skip list
                if image_name in skip_list:
                    print(f"Skipping image {image_name} as it is in the skip list.")
                    continue

                # Extract the text from the image
                extracted_text = extract_text(image_path)

                # Append the image name and extracted text to the batch data list
                batch_data.append([image_name, extracted_text])

            # Convert the batch data to a DataFrame
            df = pd.DataFrame(batch_data, columns=['Image Name', 'Extracted Text'])

            # Save the DataFrame to the CSV file (append mode)
            df.to_csv(f, header=f.tell() == 0, index=False)

            print(f"Processed and saved batch {i // batch_size + 1} of {len(image_paths) // batch_size + 1}")

            # Clear memory after each batch to prevent memory overload
            del batch_data, df
            gc.collect()  # Garbage collection to free up memory

    print(f"All batches processed and saved to {csv_file}")

# Example usage
folder_path = "/content/drive/MyDrive/Resume/Dataset/train"  # Path to your folder
csv_file = "/content/drive/MyDrive/Resume/extracted_text.csv"  # Path to save the CSV
batch_size = 100  # Adjust batch size based on your memory constraints
skip_list = a  # List of images to skip
save_text_to_csv(folder_path, csv_file, batch_size, skip_list)
```
